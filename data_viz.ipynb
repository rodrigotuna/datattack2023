## Number of occurences per district and the bottleneck

```python
# Group the occurrences by location and natureza (objective)
grouped_ocorrencias = ocorrencias.groupby(['Distrito', 'Concelho', 'Freguesia', 'Localidade', 'Natureza']).size().reset_index(name='count')

# Get the most common occurrence type per location
most_common_per_location = grouped_ocorrencias.groupby(['Distrito', 'Concelho', 'Freguesia', 'Localidade'])['Natureza', 'count'].apply(lambda x: x.loc[x['count'].idxmax()]).reset_index()

# Get the bottleneck by counting the number of occurrences per type
bottleneck = most_common_per_location.groupby('Natureza').size().reset_index(name='count').sort_values('count', ascending=False)

# Print the results
print('Most common occurrence type per location:')
print(most_common_per_location)
print('\nBottleneck in Authority’s services:')
print(bottleneck)
```

```python
import seaborn as sns


# Create a bar plot of the occurrence counts per type
plt.figure(figsize=(10, 5))
sns.barplot(data=grouped_ocorrencias, x='Natureza', y='count')
plt.title('Occurrences per Type')
plt.xlabel('Occurrence Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

# Create a pie chart of the bottleneck occurrence types
plt.figure(figsize=(10, 5))
plt.pie(bottleneck['count'], labels=bottleneck['Natureza'], autopct='%1.1f%%')
plt.title('Bottleneck Occurrence Types')
plt.show()
```

```python
# Group the occurrence data by district and natureza, and count the number of occurrences
district_natureza_counts = ocorrencias.groupby(["Distrito", "Natureza"])["Numero"].count()

# Iterate over the districts and generate a bar chart for each district, showing only the top 3 occurrence objectives
for district in district_natureza_counts.index.levels[0]:
    district_counts = district_natureza_counts[district]
    district_counts_sorted = district_counts.sort_values(ascending=False)
    top3_counts = district_counts_sorted.head(3)
    top3_counts.plot(kind="bar", title=district, rot=0)
    plt.xticks(rotation = 'vertical')
    plt.show()
```

## Comparison between true and false alarms along every hour of the day

```python
true_flags = ocorrencias.loc[(ocorrencias['EstadoOcorrencia'] != "Falso Alerta") & (ocorrencias['EstadoOcorrencia'] != "Falso Alarme")]
true_flags['Hour'] = true_flags['DataOcorrencia'].dt.hour
hour_true = true_flags['Hour'].value_counts()[true_flags['Hour'].unique()]
hour_true /= sum(hour_true)

false_flags = ocorrencias.loc[(ocorrencias['EstadoOcorrencia'] == "Falso Alerta") | (ocorrencias['EstadoOcorrencia'] == "Falso Alarme")]
false_flags['Hour'] = false_flags['DataOcorrencia'].dt.hour
hour_false = false_flags['Hour'].value_counts()[false_flags['Hour'].unique()]
hour_false /= sum(hour_false)

fig = figure(figsize=(25, 10), dpi=80)
fig.patch.set_facecolor((1.0, 1.0, 1.0))
plt.bar(false_flags['Hour'].unique()-0.2, hour_false, 0.4, label="False")
plt.bar(true_flags['Hour'].unique()+0.2, hour_true, 0.4, label="True")
plt.legend()
```

## Number of terrestrial resources and number of occurences used for each day of the year

```python
to_plot =pd.DataFrame(columns=['DiaOcorrencia'])
print(to_plot.head())
to_plot['DiaOcorrencia'] = ocorrencias['DiaOcorrencia'].unique()
to_plot['Cases'] = ocorrencias['DiaOcorrencia'].value_counts()[ocorrencias['DiaOcorrencia'].unique()].to_list()

columns = ['NumeroMeiosTerrestresEnvolvidos', 'NumeroOperacionaisTerrestresEnvolvidos', 'NumeroMeiosAereosEnvolvidos','NumeroOperacionaisAereosEnvolvidos']
for column in columns:
    meiosterrs = ocorrencias.groupby(['DiaOcorrencia']).agg({column : ['sum']})
    meiosterrs.columns = ['Sum']
    to_plot[column] = meiosterrs['Sum'].to_list()

to_plot = to_plot.set_index("DiaOcorrencia")
print(to_plot.head())
fig = figure(figsize=(100, 10), dpi=80)
fig.patch.set_facecolor((1.0, 1.0, 1.0))
plt.plot(to_plot['Cases'], label="Número Ocorrencias", color="black")
plt.plot(to_plot['NumeroMeiosTerrestresEnvolvidos'], label='Meios Terrestres', color='green')
plt.plot(to_plot['NumeroOperacionaisTerrestresEnvolvidos'], label='Opps Terrestres', color='red')
plt.legend()
plt.xticks(ticks=pd.date_range(start="2016-01-01",end="2016-12-31").to_list(),rotation='vertical')
plt.savefig("terrestrial.pdf")
```

```python
fig = figure(figsize=(100, 10), dpi=80)
fig.patch.set_facecolor((1.0, 1.0, 1.0))
plt.plot(to_plot['NumeroMeiosAereosEnvolvidos'], label='Meios Aereos', color='orange')
plt.plot(to_plot['NumeroOperacionaisAereosEnvolvidos'], label='Opps Aereos', color='purple')
plt.legend()
plt.xticks(ticks=pd.date_range(start="2016-01-01",end="2016-12-31").to_list(),rotation='vertical')
plt.savefig('aerial.pdf')
```
